## Assignment Questions

Q1. What is a parameter?

A1. A parameter is a variable in a machine learning model that the algorithm learns from the training data to make predictions, such as weights or biases.

Q2. What is correlation?

A2. Correlation is a statistical measure that indicates how strongly two variables are related to each other.
It can be positive, negative, or zero, showing the direction of the relationship.
A higher absolute value of correlation means a stronger relationship between the variables.

Q. What does negative correlation mean?

A. Negative correlation means that when one variable increases, the other variable decreases.
It shows an inverse relationship between the two variables.

Q3. Define Machine Learning. What are the main components in Machine Learning?

A3. Machine Learning is a branch of artificial intelligence where systems learn patterns from data and make predictions or decisions without being explicitly programmed.
The main components in Machine Learning are:
Data: – the information used for training and testing.
Model: – the mathematical structure that learns from data.
Algorithm: – the method used to train the model.
Prediction/Output: – the result generated by the trained model.

Q4. How does loss value help in determining whether the model is good or not?

A4. Loss value measures how far the model’s predictions are from the actual values.A lower loss value means the model is performing better, while a higher value indicates poor performance.It helps in monitoring training progress and deciding when to adjust the model or parameters.

Q5. What are continuous and categorical variables?

A5. Continuous variables are numerical values that can take any value within a range and can be measured on a scale.
Categorical variables represent distinct groups or categories and cannot be measured numerically in a meaningful way.

Q6. How do we handle categorical variables in Machine Learning? What are the common techniques?

A6. Categorical variables are handled by converting them into numerical form so that machine learning models can process them.
Common techniques include:
Label Encoding – assigning each category a unique number.
One-Hot Encoding – creating binary columns for each category.
Target Encoding – replacing categories with statistical values based on the target variab

Q7. What do you mean by training and testing a dataset?

A7. Training a dataset means using a portion of the data to teach the model and help it learn patterns.
Testing a dataset means using a separate portion of the data to evaluate how well the trained model performs on unseen data.

Q8. What is sklearn.preprocessing?

A8. sklearn.preprocessing is a module in the Scikit-learn library that provides functions to prepare and transform data before training a model. It includes tools for scaling, normalizing, encoding, and handling missing values.

Q9. What is a Test set?

A9. A test set is a portion of the dataset kept aside to evaluate the performance of a trained machine learning model.It contains data that the model has not seen during training.

Q10. How do we split data for model fitting (training and testing) in Python?

A10. In Python, data is commonly split using train_test_split() from the sklearn.model_selection module.This function divides the dataset into training and testing sets based on a specified ratio, such as 80% training and 20% testing.

Q. How do you approach a Machine Learning problem?

A. To approach a Machine Learning problem:
Understand the problem: – define goals and requirements.
Collect and prepare data: – clean, preprocess, and handle missing values.
Select a model: – choose the right algorithm for the task.
Train the model: – fit it using training data.
Evaluate and improve: – test on unseen data and fine-tune if needed.

Q11. Why do we have to perform EDA before fitting a model to the data?

A11. We perform Exploratory Data Analysis (EDA) to understand the structure, patterns, and quality of the data before modeling. EDA helps detect missing values, outliers, and data distribution issues. It guides data cleaning, feature selection, and preprocessing steps, improving model accuracy.

Q12. What is correlation?

A12. Correlation is a statistical measure that shows the strength and direction of the relationship between two variables. It helps understand how changes in one variable are associated with changes in another.

Q13. What does negative correlation mean?

A13. Negative correlation means that when one variable increases, the other variable decreases. It indicates an inverse relationship between the two variables.

Q14. How can you find correlation between variables in Python?

A14. In Python, correlation between variables can be found using the .corr() method in pandas. It calculates the correlation coefficient between columns in a DataFrame.

Q15. What is causation? Explain difference between correlation and causation with an example.

A15. Causation means that a change in one variable directly causes a change in another. Correlation only shows that two variables are related, but it does not prove one causes the other.

Q16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

A16. An optimizer is an algorithm used in machine learning to adjust the model’s parameters to minimize the loss function and improve performance.
Common types of optimizers are:
Gradient Descent: – updates parameters in the direction of the steepest decrease in loss.
Stochastic Gradient Descent (SGD): – updates parameters using one sample at a time for faster but noisier updates.
Adam: – combines momentum and adaptive learning rate for efficient and faster convergence.

Q17. What is sklearn.linear_model?

A17. sklearn.linear_model is a module in Scikit-learn that provides algorithms for linear models. It includes methods for regression, classification, and regularization, such as LinearRegression, LogisticRegression, and Ridge.

Q18. What does model.fit() do? What arguments must be given?

A18. model.fit() trains the machine learning model using the provided training data. It learns patterns by adjusting the model’s parameters based on the input features and target values.
The main arguments are:
X: – the training features (input data).
y: – the target values (output labels).

Q19. What does model.predict() do? What arguments must be given?

A19. model.predict() generates predictions from a trained machine learning model.
It takes new input data and returns the model’s estimated output.
The main argument is:
X: – the input features for which predictions are to be made.

Q21. What is feature scaling? How does it help in Machine Learning?

A21. Feature scaling is the process of adjusting the range of independent variables so they are on a similar scale. It helps machine learning models converge faster and perform better, especially those based on distance or gradient calculations.

Q22. How do we perform scaling in Python?

A22. In Python, scaling is commonly done using the sklearn.preprocessing module.
Popular methods are:
StandardScaler: – scales data to have mean 0 and standard deviation 1.
MinMaxScaler: – scales data to a fixed range, usually 0 to 1.

Q25. Explain data encoding?

A25. Data encoding is the process of converting categorical or text data into numerical form so machine learning models can process it. It ensures algorithms can interpret and learn from non-numeric information.
